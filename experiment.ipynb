```python
from pandas import read_csv
from sklearn import linear_model
import matplotlib.image as mpimage
import math
```

```python
def regressionByCoordinateAndGazeData(coordinate_data , gaze_data):

    ans_x = []
    ans_y = []
    
    diff_x = []
    diff_y = []
    
    last_ans_x = 1000
    zero_y = (coordinate_data['right_iris_y'][0] + coordinate_data['left_iris_y'][0])/2.0
    
    i = -1
    for iris_x in coordinate_data['right_iris_x']:
        i = i + 1
        if iris_x != 0 \
        and coordinate_data['right_iris_x'][i] < coordinate_data['right_eyehead_x'][i] \
        and coordinate_data['left_eyehead_x'][i] < coordinate_data['left_iris_x'][i] \
        and coordinate_data['right_iris_y'][i] < coordinate_data['right_eyehead_y'][i] \
        and coordinate_data['left_iris_y'][i] < coordinate_data['left_eyehead_y'][i] \
        and gaze_data['x'][i] != last_ans_x :
        
            ans_x.append(gaze_data['x'][i])
            ans_y.append(gaze_data['y'][i])
            
            diff_xr_i = coordinate_data['right_iris_x'][i] - coordinate_data['right_eyehead_x'][i] #minus value of < o↔>< o >
            diff_xl_i = coordinate_data['left_iris_x'][i] - coordinate_data['left_eyehead_x'][i] #< o ><↔o >
            diff_x.append(diff_xr_i + diff_xl_i)
            
            diff_y.append((coordinate_data['left_iris_y'][i] + coordinate_data['right_iris_y'][i])/2 - zero_y)
            last_ans_x = gaze_data['x'][i]

    # regreddion x
    ans_x_list = [[ans_x[i]] for i in range(len(ans_x))]
    data = np.array(ans_x_list)
    diff_x_list = [[diff_x[i]] for i in range(len(diff_x))]
    target = np.array(diff_x_list)
    
    regr_x = linear_model.LinearRegression()
    
    np.random.seed(0)
    indices = np.random.permutation(len(data))
    half = len(data)/2
    data_train = data[indices[:half]]
    data_test = data[indices[half:]]
    target_train = target[indices[:half]]
    target_test = target[indices[half:]]
    
    regr_x.fit(data_train, target_train)
    regr_x.score(data_test, target_test)
    print "(XLinearRegression:base) coef_=%s, intercept=%s" % (regr_x.coef_, regr_x.intercept_)
    
    line_x = np.arange(0, 768, 1)
    line_y = line_x * regr_x.coef_[0] + regr_x.intercept_[0]
    plot(line_x,line_y)
    scatter(ans_x, diff_x,color='r')
    ylim(-20,20)
    xlabel('x coordinate [px]')
    ylabel('x feature [px]')
    figure()
    
    # regression y
    ans_y_list = [[ans_y[i]] for i in range(len(ans_y))]
    data = np.array(ans_y_list)
    diff_y_list = [[diff_y[i]] for i in range(len(diff_y))]
    target = np.array(diff_y_list)
    
    regr_y = linear_model.LinearRegression()

    np.random.seed(0)
    indices = np.random.permutation(len(data))
    half = len(data)/2
    data_train = data[indices[:half]]
    data_test = data[indices[half:]]
    target_train = target[indices[:half]]
    target_test = target[indices[half:]]
    
    regr_y.fit(data_train, target_train)
    regr_y.score(data_test, target_test)
    print "(YLinearRegression:base) coef_=%s, intercept=%s" % (regr_y.coef_, regr_y.intercept_)
    
    scatter(ans_y, diff_y,color='r')
    line_x = np.arange(0, 1024, 1)
    line_y = line_x * regr_y.coef_[0] + regr_y.intercept_[0]
    plot(line_x,line_y)
#    ylim(-20,20)
    xlabel('y coordinate [px]')
    ylabel('y feature [px]')
    plot(line_x,line_y)
    figure()
    
    regr = [regr_x, regr_y]
    return regr
```

```python
def eyeGazeEstimation(regr, right_eyebottom_x, right_iris_x, right_eyehead_x, left_eyehead_x, left_iris_x, left_eyebottom_x, right_iris_y, left_iris_y, zero_y):
    
    regr_x = regr[0]
    regr_y = regr[1]
    
    diff_xr_i = right_iris_x - right_eyehead_x #minus value of < o↔>< o >
    diff_xl_i = left_iris_x - left_eyehead_x #< o ><↔o >
    diff_x = diff_xr_i + diff_xl_i
    
    head_y = (right_iris_y + left_iris_y)/2 - zero_y
    
    # y = a * x + b
    a_x = regr_x.coef_[0]
    b_x = regr_x.intercept_[0]
    x = (diff_x - b_x)/a_x
    
    # y = a * x + b
    a_y = regr_y.coef_[0]
    b_y = regr_y.intercept_[0]
    y = (head_y - b_y)/a_y
    eye_gaze_coordinate = [x[0],y[0]]
    
    return eye_gaze_coordinate
```
