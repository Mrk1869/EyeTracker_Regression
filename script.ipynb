```python
from pandas import read_csv
import matplotlib.image as mpimage
```

```python
data_directory_path = 'data/2013-07-02_12:58:03_LineHori_Kawaichi/'
time_data = read_csv(data_directory_path+'timestamp.csv', names=['time','x','y'], skiprows=1)
coordinate_data = read_csv(data_directory_path+'coordinate.csv', names=['time','left_iris_x','left_iris_y','left_eyehead_x','left_eyehead_y','left_eyebottom_x','left_eyebottom_y','right_iris_x','right_iris_y','right_eyehead_x','right_eyehead_y','right_eyebottom_x','right_eyebottom_y'], skiprows=1)

## show image
#time = str('%.5f' %time_data['time'][0])
#picture_path = data_directory_path + time + '.jpg'
#image = mpimage.imread(picture_path)
#plt.imshow(image)

coordinate_data[['left_iris_x','left_eyehead_x','left_eyebottom_x','right_iris_x','right_eyehead_x','right_eyebottom_x']].plot()
figure()
coordinate_data[['left_iris_y','left_eyehead_y','left_eyebottom_y','right_iris_y','right_eyehead_y','right_eyebottom_y']].plot()
figure()
```

```python
coordinate_data[['right_iris_x']].plot()
figure()
coordinate_data[['right_iris_y']].plot()
```

```python
x = []
y = []

for diffx in coordinate_data['right_iris_x']:
    if diffx != 0:
        x.append(diffx)
    
for diffy in coordinate_data['right_iris_y']:
    if diffy != 0:
        y.append(diffy)
    
plot(x, y)
```

```python
diff = coordinate_data['left_eyehead_x'] - coordinate_data['left_iris_x']
diff.plot()
figure()
diff = coordinate_data['left_eyebottom_x'] - coordinate_data['left_iris_x']
diff.plot()
```
